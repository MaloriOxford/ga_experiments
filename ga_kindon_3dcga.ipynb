```python
from kingdon import Algebra
import math

alg = Algebra(4, 1)
locals().update(alg.blades)
```

We don't work directly in the e3/e4 basis, but instead rotate it so we have two null vectors to work with (called origin and infinite)

```python
ni = e4 + e5
no = 0.5 * (e5 - e4)
nino = ni ^ no

up = lambda x: no + x + 0.5 * x * x * ni
```

Define points, lines, circles using the null basis.  

```python
point  = lambda x, y, z: no + x*e1 + y*e2 + z*e3 + 0.5*(x*x+y*y+z*z)*ni
# line   = lambda a,b,c,d: (a*e1 + b*e2 + d*ni).dual()
sphere = lambda x,y,z,r: (point(x,y,z) - r**2/2*ni).dual()
```

Distances and Angles. 

```python
dist   = lambda x,y: (2*(x.rc(y)).norm)**0.5
angle  = lambda x,y: math.acos(x.normalized().dual().rc(y.normalized().dual()))
```

Define two lines, one directly, one by wedging two points and infinity.

```python
# X = line(0,1,0,0)
l1 = lambda: p1 ^ p2 ^ ni
l2 = lambda: p2 ^ p3 ^ ni
l3 = (p1 ^ p2 ^ ni) - (p2 ^ p3 ^ ni)

X = point(0, 0, 0) ^ point(1, 0, 0) ^ ni
Y = point(0, 0, 0) ^ point(0, 1, 0) ^ ni
Z = point(0, 0, 0) ^ point(0, 0, 1) ^ ni
```

Create motors to move a point.

```python
translator = lambda x, y, z: 1 + ((x*e1 + y*e2 + z*e3) * ni)/2

t1 = translator(1, 1, 1)
t2 = translator(2, 0, -1)

# m1 = m1.normalized()
# m2 = m2.normalized()

p0 = point(0, 0, 0)
p01 = lambda: t1 * p0 * t1.reverse()
p02 = lambda: t2 * p01 * t2.reverse()
dist = lambda p1, p2: (-2 * (p1() | p2())) ** (1/2)
# dist(p01, p02)

m = [2, 3, 2]
p012 = lambda: (m[0] * p0() + m[1] * p01() + m[2] * p02()) / sum(m)
```
